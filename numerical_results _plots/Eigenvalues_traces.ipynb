```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Eigenvalues of the Prime state

```python
exact_eigen = np.load('eig_entropy_n30.npy')
exact_eigen = sorted(exact_eigen,reverse=True)

approx_eigen = np.load('analytic_eigs_n30.npy')
approx_eigen = sorted(approx_eigen,reverse=True)
```

```python
plt.plot(range(1,72),-np.log(exact_eigen)[0:71],'o')
plt.plot(range(1,72),-np.log(approx_eigen)[0:71],'o')

plt.rcParams["axes.linewidth"]=1.3
plt.ylabel('$\epsilon_k$',fontsize=12)
plt.legend(('Exact','Approximated'),loc=4)
plt.xlim([-0.3,85])
plt.grid()
plt.title('Entanglement spectrum of the Prime State')

plt.text(3,2.28,'$\phi(1)=1$',fontsize=8)
plt.text(5,3.62,'$\phi(3)=2$',fontsize=8)
plt.text(8.9,5.05,'$\phi(5)=4$',fontsize=8)
plt.text(15,5.8,'$\phi(7)=6$',fontsize=8)
plt.text(22.5,6.37,'$\phi(15)=8$',fontsize=8)
plt.text(32.5,6.8,'$\phi(11)=10$',fontsize=8)
plt.text(56.7,7.13,'$\phi(13)+\phi(21)=24$',fontsize=8)
plt.text(73,7.65,'$\phi(17)=16$',fontsize=8)
```
