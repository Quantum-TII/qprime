```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Scaling of the entanglement entropy of the Prime state

```python
primeentropy_quad_precision = [1.206187630964165157324932500000502,1.474829360680612828815295040199893,
                              2.314037122608578714575158999066184,3.190029362798827811363489469059070,
                              4.022045326650358357206221658612917,4.899349105104681172895165715584857,
                              5.787153058466823387349366983687425,6.674819075104564873723147385748711,
                              7.557377354183919577692490630357344,8.442753894802047530303350677921823,
                              9.330115611869290864096644220900359,10.21594927175403238428578158276712,
                              11.10176052571546442212334077707919,11.98755155862601156294732708934157]
                                # from n=4 to n=30 qubits
```

```python
c_pi = [primeentropy_quad_precision[i]-primeentropy_quad_precision[i-1] 
        for i in range(1,len(primeentropy_quad_precision))] # from n=4 to n=30 qubits
```

```python
gamma = [c_pi[i-1]*(i+2)-primeentropy_quad_precision[i] for i in range(1,len(primeentropy_quad_precision))] 
            # from n=4 to n=30 qubits
```

```python
ref = 3/np.pi**2
H_ref = -ref*np.log2(ref)-(1-ref)*np.log2(1-ref)
```

```python
plt.plot(range(6,32,2),c_pi,'o-',color='g')
plt.axhline(H_ref, color='k')

plt.rcParams["axes.linewidth"]=1.3
plt.xticks(range(6,32,2),range(6,32,2))
plt.xlabel('n',fontsize=12)
plt.ylabel('$c_{\pi}$',fontsize=12)
plt.ylim([0.2,1])
plt.text(28.1,0.93,r'${\bf H\left(\frac{3}{\pi^2}\right)}$')
plt.grid()
plt.title('Value of $c_{\pi}$')

sub_axes = plt.axes([.533, .25, .35, .35]) 
sub_axes.plot(range(14,32,2),c_pi[4:],'o-',color='g')
sub_axes.axhline(H_ref, color='k')
plt.xticks(range(14,32,2),range(14,32,2))
plt.text(26.8,0.8884,r'${\bf H\left(\frac{3}{\pi^2}\right)}$')
plt.ylim([0.874,0.892])
plt.grid()
```

```python
plt.plot(range(6,32,2),gamma,'o-',color='r')
plt.axhline(1+ref, color='k')

plt.rcParams["axes.linewidth"]=1.3
plt.xticks(range(6,32,2),range(6,32,2))
plt.xlabel('n',fontsize=12)
plt.ylabel('$\gamma$',fontsize=12)
plt.ylim([-0.8,1.8])
plt.text(27.3,1.44,r'${\bf 1+\frac{3}{\pi^2}}$')
plt.grid()
plt.title('Value of $\gamma$')


sub_axes = plt.axes([.533, .23, .35, .35]) 
sub_axes.plot(range(14,32,2),gamma[4:],'o-',color='r')
sub_axes.axhline(1+ref, color='k')

plt.xticks(range(14,32,2),range(14,32,2))
plt.text(26.5,1.316,r'${\bf 1+\frac{3}{\pi^2}}$')
plt.ylim([1.22,1.34])
plt.grid()
```

```python
plt.plot(range(6,32,2),np.log(abs(np.array(c_pi)-H_ref)),'o')
plt.plot(range(6,32,2),np.log(abs(np.array(gamma)-1-ref)),'o')

plt.rcParams["axes.linewidth"]=1.3
plt.xticks(range(6,32,2),range(6,34,2))
plt.legend(('$c_\pi$','$\gamma$'),fontsize=12)
plt.xlabel('n',fontsize=12)
plt.ylabel('ln $\delta$',fontsize=12)
plt.grid()
plt.title('Exponential convergence of $c_\pi$ and $\gamma$')
```
